```python
import pandas as pd
import os
import datetime


# 호텔 작업 경로 변환
os.chdir("./file_csv/hotel")

# 데이터 병합
list_hotel_2p = ['hotel_data_8_2p', 'hotel_data_9_2p', 'hotel_data_10_2p']
list_hotel_4p = ['hotel_data_8_4p', 'hotel_data_9_4p', 'hotel_data_10_4p']

data = {}
for i in list_hotel_2p:
    data[i] = pd.read_csv(f'{i}.csv')
all_data_2p = pd.concat(data.values(), ignore_index=True)

data = {}
for i in list_hotel_4p:
    data[i] = pd.read_csv(f'{i}.csv')
all_data_4p = pd.concat(data.values(), ignore_index=True)

all_data_2p['person'] = '2인'
all_data_4p['person'] = '4인'
all_data = pd.concat([all_data_2p, all_data_4p], axis = 0)


def hotel_EDA(file):

    # 결측치 및 중복값 삭제
    df = file.dropna()
    df = df.drop_duplicates()
    # 결측치 및 중복값 삭제
    df = file.dropna()
    df = df.drop_duplicates()

    # 요일 컬럼 생성
    df['date'] = pd.to_datetime(df['date'])
    dateDict = {0:"월요일", 1:"화요일", 2:"수요일", 3:"목요일", 4:"금요일", 5:"토요일", 6:"일요일"}   

    def day_output(date):
        return dateDict[date.weekday()]
    
    df['day'] = df['date'].apply(day_output)

    # 주말 / 공휴일 컬럼 생성
    holiday_list = ['2023-08-15', '2023-09-28', '2023-09-29', '2023-09-30', '2023-10-03', '2023-10-09']
    weekends_list = ['금요일', '토요일', '일요일']

    def day_off(row):
        date = pd.to_datetime(row['date']).strftime('%Y-%m-%d')  # 날짜 형식을 통일하기 위해 문자열로 변환
        if date in holiday_list:
            return '공휴일'
        elif row['day'] in weekends_list:
            return '주말'
        else:
            return '평일'

    df['holiday'] = df.apply(day_off, axis=1)
    
    # csv 파일 변환
    df.to_csv("hotel_EDA.csv", index=False, encoding="utf-8-sig", mode="w")

    return df

df = hotel_EDA(all_data)
print(df)
```
