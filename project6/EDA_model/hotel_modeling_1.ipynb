### 주소 (location)

```python
# 주소가 너무 길음 >> 시, 동으로 구분

def city_find(address):
    address_parts = address.split()[1]
    return address_parts
df['city'] = df['location'].apply(city_find)

def city_find2(address):
    address_parts = address.split()[2]
    return address_parts
df['city_details'] = df['location'].apply(city_find2)
```

### review_count의 카디널리티가 매우 높음

```python
# 500미만 / 500이상으로 분류
```

```python
# 주소가 너무 길음 >> 시, 동으로 구분

def review_group(review_count):
    if review_count < 500:
        return '500 미만'
    else:
        return '500 이상'
df['review_group'] = df['review_count'].apply(review_group)
df
```

### room_type의 카디널리티가 매우 높음

```python
# 2인과 4인 데이터를 따로 이용하여 room_type를 없애고, 가격은 같은 호텔 내에서 평균가로 결정
```

#### 2인

```python
# 빈 표 만들기 (df_)
df_ = df.drop(['platform', 'level', 'score', 'review_count', 'room_type', 'location', 'link', 'day', 'holiday', 'review_group', 'day_diff', 'person'], axis=1)
df_ = df_.drop(range(0, 201172), axis=0)
df_

# date 리스트 (중복 x)
date_sort = df_2p['date'].sort_values(ignore_index=True)
date_sort_list = [date_sort[i] for i in range(len(date_sort))]  # 이름 >> 리스트화
date_list = list(dict.fromkeys(date_sort_list))  # 중복 제거

# name 리스트 (중복 x)
name_sort = df_2p['name'].sort_values(ignore_index=True)
name_sort_list = [name_sort[i] for i in range(len(name_sort))]  # 이름 >> 리스트화
name_list = list(dict.fromkeys(name_sort_list))  # 중복 제거

dict_ = {}
def mean_by_hotel():
    for x in range(len(date_list)):
        for y in range(len(name_list)):
            price = df_2p.query(f'(name=="{name_list[y]}") and (date=="{date_list[x]}")')['price'].mean()
            dict_[(100000*x)+(y)] = {'date': date_list[x], 'name': name_list[y], 'price': price}
    new_df = pd.DataFrame.from_dict(dict_, "index")
    new_df['person'] = '2인'
    return new_df

df_2p_ = mean_by_hotel()
```

```python
df2 = df_2p_.dropna().reset_index(drop=True)

df_nlr = df[['name', 'level', 'review_count', 'review_group', 'score', 'city', 'city_details']].drop_duplicates().reset_index(drop = True)
df_nlr = df_nlr.loc[df_nlr[['name']].drop_duplicates().index].reset_index(drop = True)

# merge (새로운 데이터에 level, review_count 추가)
df3 = pd.merge(df2, df_nlr, how = 'inner', on = 'name')
df3
```

```python
df3 = df3.sort_values(by='date', ascending=True)
df3 = df3.reset_index(drop=True)
df3['price'] = df3['price'].apply(lambda x: round(x))
df3
```

```python
# csv 파일 변환
df3.to_csv("hotel_EDA_2p_2.csv", index=False, encoding="utf-8-sig", mode="w")
```
